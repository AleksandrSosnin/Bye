```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/kc_house_data.csv', delimiter=',', encoding='utf-8')

if len(df.columns) == 1:
    df = pd.read_csv('your_file.csv', delimiter=',', encoding='utf-8', engine='python')

print(df.head())
print(df.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Изучение стоимости недвижимости
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['sqft_living'], bins=30, kde=True)
plt.title('Распределение квадратуры жилой площади')
plt.xlabel('Квадратура (кв. футы)')
plt.ylabel('Количество')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['yr_built'], bins=30, kde=True)
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество')
plt.show()
```

```python
# 2.1. Изучите распределение домов от наличия вида на набережную
plt.figure(figsize=(10, 6))
sns.countplot(x='waterfront', data=df)
plt.title('Распределение домов по наличию вида на набережную')
plt.xlabel('Вид на набережную (0 - нет, 1 - есть)')
plt.ylabel('Количество')
plt.show()

# Выводы по наличию вида на набережную
print("Количество домов без вида на набережную:", df['waterfront'].value_counts()[0])
print("Количество домов с видом на набережную:", df['waterfront'].value_counts()[1])
```

```python
# 2.2. Изучите распределение этажей домов
plt.figure(figsize=(10, 6))
sns.countplot(x='floors', data=df)
plt.title('Распределение этажей домов')
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов')
plt.show()
```

```python
# 2.3. Изучите распределение состояния домов
plt.figure(figsize=(10, 6))
sns.countplot(x='condition', data=df)
plt.title('Распределение состояния домов')
plt.xlabel('Состояние дома')
plt.ylabel('Количество домов')
plt.show()
```

```python
# 1. Распределение стоимости недвижимости
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество')
plt.show()
print("Средняя стоимость недвижимости:", df['price'].mean())
print("Медианная стоимость недвижимости:", df['price'].median())
print("Максимальная стоимость недвижимости:", df['price'].max())
```

```python
# 2. Влияние жилой площади на стоимость
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sqft_living', y='price', data=df)
plt.title('Влияние жилой площади на стоимость')
plt.xlabel('Жилая площадь (кв. футы)')
plt.ylabel('Стоимость')
plt.show()
```

```python
# 3. Влияние количества спален на стоимость
plt.figure(figsize=(10, 6))
sns.boxplot(x='bedrooms', y='price', data=df)
plt.title('Влияние количества спален на стоимость')
plt.xlabel('Количество спален')
plt.ylabel('Стоимость')
plt.show()
```

```python
# 4. Влияние количества ванных комнат на стоимость
plt.figure(figsize=(10, 6))
sns.boxplot(x='bathrooms', y='price', data=df)
plt.title('Влияние количества ванных комнат на стоимость')
plt.xlabel('Количество ванных комнат')
plt.ylabel('Стоимость')
plt.show()
```

```python
# 5. Влияние года постройки на стоимость
plt.figure(figsize=(10, 6))
sns.scatterplot(x='yr_built', y='price', data=df)
plt.title('Влияние года постройки на стоимость')
plt.xlabel('Год постройки')
plt.ylabel('Стоимость')
plt.show()
```

```python
# 6. Влияние состояния дома на стоимость
plt.figure(figsize=(10, 6))
sns.boxplot(x='condition', y='price', data=df)
plt.title('Влияние состояния дома на стоимость')
plt.xlabel('Состояние дома')
plt.ylabel('Стоимость')
plt.show()
```

```python
# 7. Влияние наличия вида на набережную на стоимость
plt.figure(figsize=(10, 6))
sns.boxplot(x='waterfront', y='price', data=df)
plt.title('Влияние наличия вида на набережную на стоимость')
plt.xlabel('Вид на набережную (0 - нет, 1 - есть)')
plt.ylabel('Стоимость')
plt.show()
```


### 1. Распределение стоимости недвижимости
**График:** Гистограмма стоимости недвижимости

**Вывод:**
- Средняя стоимость недвижимости составляет около **...** долларов, с медианной стоимостью в **...** долларов и максимальной ценой достигающей **...** долларов.
- Это показывает, что рынок недвижимости насыщен вариантами для любого бюджета, от уютных домов до роскошных особняков.

### 2. Влияние жилой площади на стоимость
**График:** График рассеяния (Scatter plot) "Жилая площадь vs. Стоимость"

**Вывод:**
- Большая жилая площадь — больший потенциал для роста цены.
- Вложение в просторную жилую площадь — это инвестиция, которая всегда окупается.
- Если вы хотите повысить ценность своего дома, начните с расширения его пространства.

### 3. Влияние количества спален на стоимость
**График:** Боксплот (Boxplot) "Количество спален vs. Стоимость"

**Вывод:**
- Количество спален имеет прямое влияние на стоимость дома.
- Дома с 3-4 спальнями пользуются особой популярностью.
- Если вы планируете продавать дом, добавление еще одной спальни может существенно увеличить его рыночную стоимость.

### 4. Влияние количества ванных комнат на стоимость
**График:** Боксплот (Boxplot) "Количество ванных комнат vs. Стоимость"

**Вывод:**
- Больше ванных комнат — больше удобства для всех членов семьи, и это, безусловно, увеличивает стоимость дома.
- Дома с 2-3 ванными комнатами, как правило, стоят дороже.
- Обеспечьте комфорт своим близким и увеличьте стоимость своего дома с дополнительной ванной комнатой.

### 5. Влияние года постройки на стоимость
**График:** График рассеяния (Scatter plot) "Год постройки vs. Стоимость"

**Вывод:**
- Новые дома, построенные после 2000 года, пользуются большим спросом и имеют более высокую стоимость.
- Современные дома предлагают лучшие стандарты жизни и энергоэффективность.
- Если ваш дом старше, подумайте о модернизации, чтобы увеличить его привлекательность и стоимость.

### 6. Влияние состояния дома на стоимость
**График:** Боксплот (Boxplot) "Состояние дома vs. Стоимость"

**Вывод:**
- Хорошее состояние дома — залог его высокой стоимости.
- Дома в отличном состоянии (4-5) ценятся значительно выше.
- Регулярное обслуживание и обновление вашего дома не только создают комфорт, но и повышают его рыночную стоимость.

### 7. Влияние наличия вида на набережную на стоимость
**График:** Боксплот (Boxplot) "Вид на набережную vs. Стоимость"

**Вывод:**
- Вид на набережную — это не просто пейзаж за окном, это престиж и высокая стоимость вашей недвижимости.
- Дома с видом на воду стоят значительно дороже.
- Хотите продать дом по максимальной цене? Подумайте о видах и расположении!
